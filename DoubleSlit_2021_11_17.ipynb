```python
def I_envelope (x, lam = 0.2e-6, a = None, D = 39.0e-6):
    if a is None:
        a = 2. * lam
    return (
        np.sin(
            (np.pi * a / lam)*np.sin(np.arctan(x / D))
        ) / 
        (
            (np.pi * a / lam)*np.sin(np.arctan(x / D))
        )
    )**2
```

```python
def I_exp (x, d = None, D = 19.5e-6, lam = 0.2e-6):
    if d is None:
        d = lam*6
    return(
        4 * np.cos(
            (d * np.pi * x)
            /
            (lam * D)
        )**2
    )
```

```python
def max_loc(d = 19.5e-6, lam = 0.2e-6, slit_d = None):
    if slit_d is None:
        slit_d = lam*6
    x_loc = []
    for n in [0, 1, 2]:
        x_loc.append( d * np.tan(np.arcsin(n * lam / slit_d)))
        x_loc.append( -d * np.tan(np.arcsin(n * lam / slit_d)))
    return x_loc
    
max_loc()
```

```python
I_0 = np.max(S)

f=plt.figure()
ax=f.gca()
ax.plot(x, S, label='measured')
ax.plot(x, I_0/4 * I_exp(x), label='expected (theory)')
ax.vlines(x=max_loc(), ymin=0, ymax=I_0, label='Maxima (theory)', color='grey')
plt.legend()
```
