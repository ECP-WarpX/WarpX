```python
offset = np.min(S)
S=S-offset
```

```python
x1=df1.query('`[0]step()` == 5300')['[2]part_x_lev0-(m)']
Ex1=df1.query('`[0]step()` == 5300')['[5]part_Ex_lev0-(V/m)']
```

```python
def I_envelope (x, lam = 0.2e-6, a = None, D = 23e-6):
    if a is None:
        a = 2 * lam
    return (
        np.sin(
            (np.pi * a / lam)*np.sin(np.arctan(x / D))
        ) / 
        (
            (np.pi * a / lam)*np.sin(np.arctan(x / D))
        )
    )**2
```

```python
def I_exp (x, d = None, D = 23e-6, lam = 0.2e-6):
    if d is None:
        d = lam*6
    return(
        4 * np.cos(
            (d * np.pi * x)
            /
            (lam * D)
        )**2
    )
```

```python
def max_loc(D = 23e-6, lam = 0.2e-6, slit_d = None):
    if slit_d is None:
        slit_d = lam*6
    x_loc = []
    for n in [0, 1]:
        x_loc.append( D * np.tan(np.arcsin(n * lam / slit_d)))
        x_loc.append( -D * np.tan(np.arcsin(n * lam / slit_d)))
    return x_loc
    
max_loc()
```

```python
I_0 = np.max(S)

f=plt.figure()
ax=f.gca()
ax.plot(x, S, label='measured')
ax.plot(x, I_0 * I_envelope(x), label='expected envelope')
#ax.plot(x, I_0/4 * I_exp(x), label='expected (theory)')

ax.vlines(x=max_loc(), ymin=0, ymax=I_0, label='Maxima (theory)', color='grey')
plt.legend()
```

```python
def perc (T = 10.05e-6, Z1 = -2.5e-6, Z2 = 24e-6):
    return (
        (
            T - (Z2 - Z1)/2
        )
        /
        (
            Z2 - (Z2 - Z1)/2
        )
    )
    
perc()
```

```python
lineEx , lineEx_info = ts.get_field(field='E', coord='x', slice_across = 'z', slice_relative_position = perc(), iteration = 5300, plot=True, label='OpenPMD')
xpmd = np.linspace(-10e-6, 10e-6, np.size(lineEx))
```
