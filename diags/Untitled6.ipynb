```python
import yt
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
```

```python
ds = yt.load('./diag101000')
ds.field_list
Nz = ds.domain_dimensions[0]
ad0 = ds.covering_grid(level=0, left_edge=ds.domain_left_edge, dims=ds.domain_dimensions)
z = np.linspace(ds.domain_left_edge[0],ds.domain_right_edge[0],Nz)

Ez_array = ad0[('boxlib','Ez')].to_ndarray()
Ez = Ez_array.reshape((Nz,))

Ey_array = ad0[('boxlib','Ey')].to_ndarray()
Ey = Ey_array.reshape((Nz,))

rho_array = ad0[('boxlib','rho')].to_ndarray()
rho = rho_array.reshape((Nz,))


plt.figure()
plt.plot(z*1e6,Ez/1e9,'b')
#plt.ylim([0,0.2])
fontprop = fm.FontProperties(size=12) #Using different fonts
plt.xlabel("z, $\mu m$", fontproperties=fontprop, fontsize=16)            #Formatting X axis label
plt.ylabel("Ez, GV/m", fontproperties=fontprop, fontsize=16)  #Formatting Y axis label
plt.xticks(fontproperties=fontprop,fontsize=13)                            #Formatting the x axis ticks
plt.yticks(fontproperties=fontprop, fontsize=13)                        #Formatting the Y axis ticks
plt.savefig('Ez.png', dpi=300, bbox_inches='tight')

plt.figure()
plt.plot(z*1e6,Ey/1e12,'b')
#plt.ylim([0,0.2])
fontprop = fm.FontProperties(size=12) #Using different fonts
plt.xlabel("z, $\mu m$", fontproperties=fontprop, fontsize=16)            #Formatting X axis label
plt.ylabel("Ey, TV/m", fontproperties=fontprop, fontsize=16)  #Formatting Y axis label
plt.xticks(fontproperties=fontprop,fontsize=13)                            #Formatting the x axis ticks
plt.yticks(fontproperties=fontprop, fontsize=13)                        #Formatting the Y axis ticks
plt.savefig('Ey.png', dpi=300, bbox_inches='tight')

plt.figure()
plt.plot(z*1e6,rho,'b')
#plt.ylim([0,0.2])
fontprop = fm.FontProperties(size=12) #Using different fonts
plt.xlabel("z, $\mu m$", fontproperties=fontprop, fontsize=16)            #Formatting X axis label
plt.ylabel("rho", fontproperties=fontprop, fontsize=16)  #Formatting Y axis label
plt.xticks(fontproperties=fontprop,fontsize=13)                            #Formatting the x axis ticks
plt.yticks(fontproperties=fontprop, fontsize=13)                        #Formatting the Y axis ticks
plt.savefig('rho.png', dpi=300, bbox_inches='tight')
```
